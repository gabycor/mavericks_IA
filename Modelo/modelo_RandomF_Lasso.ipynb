# Librerías

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import Lasso
```

# Modelo Random Forest

## Label DF

```python
# Data Split
X = df_label.drop("SalePriceLog", axis = 1)
y = df_label["SalePriceLog"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
params = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],  # Agregar max_features
    'min_weight_fraction_leaf': [0.0, 0.1, 0.2]  # Agregar min_weight_fraction_leaf
}
```

```python
# Random Forest Regressor
rfr_label = RandomForestRegressor(random_state=42)

# Uso de validación cruzada para ajustar los hiperparámetros en el conjunto de entrenamiento
search_rfr_label = GridSearchCV(rfr_label, params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Ajuste de datos del entrenamiento
search_rfr_label.fit(X_train, y_train)
```

```python
# Obtiene los mejores parámetros
best_params_rfr = search_rfr_label.best_params_

print("Mejores parámetros:", best_params_rfr)
print("Best score:", search_rfr_label.best_score_)
```

```python
# Random Forest Regressor con los mejores hiperparámetros
randomforest_LabelBest = RandomForestRegressor(
    max_depth=best_params_rfr.get("max_depth"),
    min_samples_leaf=best_params_rfr.get("min_samples_leaf"),
    max_features=best_params_rfr.get("max_features"),
    min_weight_fraction_leaf=best_params_rfr.get("min_weight_fraction_leaf"),
    n_estimators=best_params_rfr.get("n_estimators"),
    min_samples_split=best_params_rfr.get("min_samples_split"),
    random_state=0
)
```

```python
randomforest_LabelBest.fit(X_train,y_train)
```

```python
y_pred = np.exp(randomforest_LabelBest.predict(X_test))
y_test = np.exp(y_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Modelo Lasso

## Label DF

```python
# Data Split
X = df_label.drop("SalePriceLog", axis = 1)
y = df_label["SalePriceLog"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Parámetros por analizar
params = {'alpha': [0.01, 0.1, 1.0], 'fit_intercept': [True, False]}

# Modelo LDA
lasso_model_label = Lasso()
```

```python
# Uso de validación cruzada para ajustar los hiperparámetros en el conjunto de entrenamiento
search_lasso_label = GridSearchCV(lasso_model_label, params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
# Ajuste de datos del entrenamiento
search_lasso_label.fit(X_train, y_train)
```

```python
# Obtiene los mejores parámetros
best_params_lasso_label = search_lasso_label.best_params_

print("Mejores parámetros:", best_params_lasso_label)
print("Best score:", search_lasso_label.best_score_)
```

```python
# Modelo Lasso con los mejores hiperparámetros
lasso_best_model = Lasso(
    alpha=best_params_lasso_label['alpha'],
    fit_intercept=best_params_lasso_label['fit_intercept']
)
```

```python
lasso_best_model.fit(X_train, y_train)

# Predicciones en el conjunto de prueba
y_pred = lasso_best_model.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
