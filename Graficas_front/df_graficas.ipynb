# DF final

```python
filtrado = df[['MSZoning', 'LotArea', 'OverallCond', 'YearBuilt', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenQual', 'GarageCars', 'PoolQC', 'SalePrice']]
```

```python
display(filtrado)
```

```python
filtrado['MSZoning'] = filtrado['MSZoning'].replace({'A': 'Agricultura', 'C (all)': 'Comercial', 'FV': 'Residencial Pueblo Flotante', 'I': 'Industrial', 'RH': 'Residencial Alta Densidad', 'RL': 'Residencial Baja Densidad', 'RP': 'Parque Residencial Baja Densidad', 'RM': 'Residencial Densidad Media'})
filtrado['KitchenQual'] = filtrado['KitchenQual'].replace({'Ex': 'Excelente', 'Gd': 'Bueno', 'TA': 'Regular', 'Fa': 'Malo'})
filtrado['PoolQC'] = filtrado['KitchenQual'].replace({'nan': 'Sin alberca', 'Ex': 'Excelente', 'Gd': 'Bueno', 'Fa': 'Malo'})
```

```python
# Rangos de años
rangos = list(range(1870, 2011, 5))
filtrado['YearBuilt'] = pd.cut(filtrado['YearBuilt'], bins=rangos, labels=[f'{i}-{i+4}' for i in rangos[:-1]])
```

```python
display(filtrado)
```

```python
filtrado.sort_values(by='LotArea', inplace=True)
unique_count = filtrado['LotArea'].nunique()
```

```python
fg = filtrado.sort_values(by='LotArea', inplace=True)
fg.unique().count()
```

```python
rangos = [0, 500, 1000, 5000, 10000, 20000, 50000, 100000, 215245]
etiquetas = ['0-500', '500-1000', '1000-5000', '5000-10000', '10000-20000', '20000-50000', '50000-100000', '100000-215245']

filtrado['LotArea'] = pd.cut(filtrado['LotArea'], bins=rangos, labels=etiquetas)
```

```python
filtrado['N_Bath'] = filtrado['FullBath'] + filtrado['HalfBath']
```

```python
filtrado.drop(columns=['FullBath', 'HalfBath'], inplace=True)
```
